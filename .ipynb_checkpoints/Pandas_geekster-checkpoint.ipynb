```python
a = [1,2,3,4,5]
df = pd.Series(a)
df
# 0,1,2,3,4, are index/labels
```

```python
# We can change the index also 
df = pd.Series(a,index=['a','b','c','d','e'])
df
```

```python
print(df[1],df['b'])
# After changing the index we can call using both index and lables we can also have duplicate index
```

```python
# Dataframe is a combinaton of series
d = {
    'ff' :[1,2,3,4,5],
    'cc' :[6,7,8,9,10]
}
df = pd.DataFrame(d,index=['a','b','c','d','e'])
df
```

```python
# Accessing row in dataframe we use loc function 
# loc => it is basically used for row name
df.loc['a':'c','ff':]
```

```python
# iloc is also used for accessing the data but it use integer value/index name index doesn't work
# iloc => integer location
df.iloc[0:2,0:1]
```

# Creating new column

```python
def age_(age):
    if age<20:
        return 'minor'
    if age>=20 and age<40:
        return 'adult'
    elif age>40:
        return 'old'
```

```python
df["Age_comb"]=df['Age'].apply(age_)
```
